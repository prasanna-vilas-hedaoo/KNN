```python
import pandas as pd
```

```python
data = pd.read_csv("/content/glass.csv")
```

```python
data
```

```python
list(data)
```

```python
## split data
data.info()
```

```python
x = data.iloc[0:, :-1]
x
```

```python
y = data.iloc[:, -1]
y
```

```python
x.head()
```

```python
x.corr()
```

```python
## data transformation
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()   ## range -3 to +3
```

```python
s_tr = ss.fit_transform(x)
```

```python
s_tr
```

```python
s_df = pd.DataFrame(s_tr)
```

```python
s_df
```

```python
s_df.columns = list(x)
```

```python
s_df.head()
```

```python
## data partition
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(s_df,y, test_size = 0.30)
```

```python
## fit the model
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
## model predicted values
y_pred_train = knn.predict(x_train)
y_pred_test = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc1 = accuracy_score(y_train,y_pred_train)
acc2 = accuracy_score(y_test,y_pred_test)
```

```python
print("training accuracy score:", acc1.round(2))
print("test accuracy score:", acc2.round(2))
```

```python
## trying other methods to increase accuracy
```

```python
## cross validation
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
```

```python
import numpy as np
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
training_accuracy = []
test_accuracy = []

for i in range(1,101):
    x_train,x_test,y_train,y_test = train_test_split(s_df,y, test_size=0.30,random_state=i)
    knn.fit(x_train,y_train)
    y_pred_train =  knn.predict(x_train)
    y_pred_test  =  knn.predict(x_test)
    training_accuracy.append(accuracy_score(y_train,y_pred_train))
    test_accuracy.append(accuracy_score(y_test,y_pred_test))


print("cross validation training result:",np.mean(training_accuracy).round(2))
print("cross validation test result:",np.mean(test_accuracy).round(2))
```
